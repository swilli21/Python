```python
import yfinance as yf
import pandas as pd
```

```python
#create an object that will allow us to access functions to extract data; Ticker module
#Provide the ticker symbol for the stock, here the company is Apple and the ticker symbol is AAPL.
apple = yf.Ticker("AAPL")
```

```python
# 'info' we can extract information about the stock as a Python dictionary
apple_info=apple.info
apple_info
```

```python
# get the 'country' using the key country
apple_info['country']
```

```python
#The prices of these shares fluctuate over time. Using the history() method we can get the share price of the stock over a certain period of time
#Using the period parameter we can set how far back from the present to get data. The options for period are 1 day (1d), 5d, 1 month (1mo) , 3mo, 6mo, 1 year (1y), 2y, 5y, 10y, ytd, and max.
apple_share_price_data = apple.history(period="max")
```

```python
#The format that the data is returned in is a Pandas DataFrame. With the Date as the index the share Open, High, Low, Close, Volume, and Stock Splits are given for each day.
apple_share_price_data.head()
```

```python
#Reset the index of the DataFrame with the reset_index function. We also set the inplace paramter to True so the change takes place to the DataFrame itself.
apple_share_price_data.reset_index(inplace=True)
```

```python
#Plot the Open price against the Date
apple_share_price_data.plot(x="Date", y="Open")
```

```python
# Using the variable dividends we can get a dataframe of the data. The period of the data is given by the period defined in the 'history` function.
apple.dividends
```

```python
#plot the dividends overtime
apple.dividends.plot()
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#Define the url and load it into a variable by calling the get function
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/netflix_data_webpage.html"

data  = requests.get(url).text
```

```python
#Parse the text into html using beautiful_soup
soup = BeautifulSoup(data, 'html5lib')
```

```python
#Turn the html table into a pandas dataframe
netflix_data = pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close", "Volume"])

# First we isolate the body of the table which contains all the information
# Then we loop through each row and find all the column values for each row
for row in soup.find("tbody").find_all('tr'):
    col = row.find_all("td")
    date = col[0].text
    Open = col[1].text
    high = col[2].text
    low = col[3].text
    close = col[4].text
    adj_close = col[5].text
    volume = col[6].text
    
    # Finally we append the data of each row to the table
    netflix_data = netflix_data.append({"Date":date, "Open":Open, "High":high, "Low":low, "Close":close, "Adj Close":adj_close, "Volume":volume}, ignore_index=True)    
```

```python
#Print the dataframe first 5 rows
netflix_data.head()
```

```python
 # pandas read_html function using the url. Note: this is the same as calling the head function
read_html_pandas_data = pd.read_html(url)
```

```python
#There is only one table on the page, we just take the first table in the list returned
netflix_dataframe = read_html_pandas_data[0]

netflix_dataframe.head()
```
